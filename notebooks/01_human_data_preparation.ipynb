# Import Data

```python
import os
import json
import pandas as pd
```

```python
# 최상위 폴더 경로
base_path = "../data/raw/human"

# 폴더별 데이터프레임 저장
folder_dfs = {}

# TL 폴더 목록
folder_list = sorted([f for f in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, f))])

print(f"총 {len(folder_list)}개 폴더 처리 시작")

for folder_idx, folder_name in enumerate(folder_list, start=1):
    folder_path = os.path.join(base_path, folder_name)
    texts = []

    json_files = sorted([
        f for f in os.listdir(folder_path)
        if f.endswith(".json") and "TEXT" in f
    ])

    print(f"\n[{folder_idx}/{len(folder_list)}] 폴더 시작: {folder_name} | JSON 파일 수: {len(json_files)}")

    for file_idx, file_name in enumerate(json_files, start=1):
        file_path = os.path.join(folder_path, file_name)

        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)

            file_sentence_count = 0

            for paragraph in data.get("paragraphs", []):
                for sentence in paragraph.get("sentences", []):
                    text = sentence.get("text", "")
                    if text:
                        texts.append(text)
                        file_sentence_count += 1

            # 파일 단위 진행 로그
            if file_idx % 5 == 0 or file_idx == len(json_files):
                print(
                    f"  - 파일 진행: {file_idx}/{len(json_files)} "
                    f"({file_name}) | 현재 누적 문장 수: {len(texts):,}"
                )

        except Exception as e:
            print(f"  ! 오류 발생: {file_name} | {e}")

    # 폴더별 DataFrame 생성
    df = pd.DataFrame(texts, columns=["text"])
    folder_dfs[folder_name] = df

    print(f"[완료] {folder_name} | 총 문장 수: {len(df):,}")

print("\n모든 폴더 처리 완료")
```

```python
import pandas as pd
import numpy as np
import re

# =========================================================
# 0. 기본 설정
# =========================================================

TARGET_N = 10_000
RANDOM_STATE = 42

topic_map = {
    "320": ("경제학", df_320.copy()),
    "360": ("법률·법학", df_360.copy()),
    "370": ("교육학", df_370.copy()),
    "400": ("자연과학", df_400.copy()),
    "440": ("천문학", df_440.copy()),
}

# 문장 길이 구간(어절 수 기준)
BINS = [5, 10, 15, 20, 30, 51]   # 5~9 / 10~14 / 15~19 / 20~29 / 30~50
LABELS = ["5-9", "10-14", "15-19", "20-29", "30-50"]
```

```python
# =========================================================
# 1. 전처리 함수
# =========================================================

def normalize_text(text):
    """
    기본 텍스트 정규화
    """
    if pd.isna(text):
        return np.nan
    
    text = str(text).strip()
    
    # 공백 정리
    text = re.sub(r"\s+", " ", text)
    
    # 특수한 따옴표/기호 정리
    text = text.replace("“", '"').replace("”", '"').replace("’", "'").replace("‘", "'")
    
    return text


def is_valid_sentence(text, min_words=5, max_words=50):
    """
    정상 문장 여부 판별
    """
    if pd.isna(text):
        return False
    
    text = str(text).strip()
    
    # 너무 짧은 문자열 제거
    if len(text) < 10:
        return False
    
    # 한글 포함 여부
    if not re.search(r"[가-힣]", text):
        return False
    
    # 어절 수 기준
    word_count = len(text.split())
    if word_count < min_words or word_count > max_words:
        return False
    
    # 한글 비율이 너무 낮은 경우 제거
    hangul_chars = re.findall(r"[가-힣]", text)
    if len(hangul_chars) / max(len(text), 1) < 0.3:
        return False
    
    # 메타성/목차성/항목성 텍스트 제거
    meta_patterns = [
        r"^제\s*\d+\s*장",          # 제1장
        r"^제\s*\d+\s*절",          # 제1절
        r"^\d+\.$",                 # 1.
        r"^\d+\)",                  # 1)
        r"^부록",                   # 부록
        r"^참고문헌",               # 참고문헌
        r"^목차",                   # 목차
        r"^저자",                   # 저자
        r"^출판사",                 # 출판사
        r"^ISBN",                  # ISBN
        r"^머리말",                 # 머리말
        r"^서문",                   # 서문
        r"^차례",                   # 차례
    ]
    for pattern in meta_patterns:
        if re.search(pattern, text):
            return False
    
    # 기호/숫자 비중이 지나치게 높은 경우 제거
    special_count = len(re.findall(r"[^가-힣a-zA-Z0-9\s\.\,\?\!]", text))
    if special_count / max(len(text), 1) > 0.3:
        return False

    return True


def add_length_features(df, text_col="text"):
    """
    어절 수 및 길이 구간 추가
    """
    df = df.copy()
    df["word_count"] = df[text_col].str.split().str.len()
    df["length_bin"] = pd.cut(
        df["word_count"],
        bins=BINS,
        labels=LABELS,
        right=False,
        include_lowest=True
    )
    return df
```

```python
# =========================================================
# 2. 정제 함수
# =========================================================

def clean_dataframe(df, text_col="text"):
    """
    데이터프레임 정제
    """
    out = df.copy()
    
    # text 컬럼만 사용
    out = out[[text_col]].copy()
    
    # 문자열 정규화
    out[text_col] = out[text_col].apply(normalize_text)
    
    # 결측 제거
    out = out.dropna(subset=[text_col])
    
    # 빈 문자열 제거
    out = out[out[text_col].str.strip() != ""]
    
    # 유효 문장 필터링
    out = out[out[text_col].apply(is_valid_sentence)]
    
    # 중복 제거
    out = out.drop_duplicates(subset=[text_col]).reset_index(drop=True)
    
    # 길이 정보 추가
    out = add_length_features(out, text_col=text_col)
    
    # 길이 구간 없는 것 제거
    out = out.dropna(subset=["length_bin"]).reset_index(drop=True)
    
    return out
```

```python
# =========================================================
# 3. 길이 분포 유지 층화 샘플링
# =========================================================

def stratified_sample_by_length(df, target_n=100_000, text_col="text", random_state=42):
    """
    문장 길이 구간 분포를 유지하며 층화 샘플링
    """
    df = df.copy()
    
    if len(df) < target_n:
        raise ValueError(f"표본 수가 부족하다. 현재 {len(df):,}개, 목표 {target_n:,}개이다.")
    
    # 각 구간별 비율 계산
    bin_counts = df["length_bin"].value_counts().sort_index()
    bin_ratios = bin_counts / bin_counts.sum()
    
    # 목표 샘플 수 계산
    target_counts = (bin_ratios * target_n).round().astype(int)
    
    # 반올림 오차 보정
    diff = target_n - target_counts.sum()
    if diff != 0:
        # 표본이 많은 구간부터 보정
        sorted_bins = bin_counts.sort_values(ascending=False).index.tolist()
        for i in range(abs(diff)):
            b = sorted_bins[i % len(sorted_bins)]
            target_counts[b] += 1 if diff > 0 else -1
    
    sampled_parts = []
    
    for bin_label, n in target_counts.items():
        subset = df[df["length_bin"] == bin_label]
        if len(subset) < n:
            # 해당 구간이 부족하면 가능한 만큼만 먼저 추출
            sampled = subset.sample(n=len(subset), random_state=random_state)
        else:
            sampled = subset.sample(n=n, random_state=random_state)
        sampled_parts.append(sampled)
    
    sampled_df = pd.concat(sampled_parts, axis=0).reset_index(drop=True)
    
    # 부족분 보충
    shortage = target_n - len(sampled_df)
    if shortage > 0:
        remain = df[~df[text_col].isin(sampled_df[text_col])]
        extra = remain.sample(n=shortage, random_state=random_state)
        sampled_df = pd.concat([sampled_df, extra], axis=0).reset_index(drop=True)
    
    # 최종 셔플
    sampled_df = sampled_df.sample(frac=1, random_state=random_state).reset_index(drop=True)
    
    return sampled_df
```

```python
# =========================================================
# 4. 통계 확인 함수
# =========================================================

def summarize_df(df, topic_name, text_col="text"):
    summary = {
        "주제": topic_name,
        "문장 수": len(df),
        "총 어절 수": int(df["word_count"].sum()),
        "평균 어절 수": round(df["word_count"].mean(), 2),
        "중앙값 어절 수": int(df["word_count"].median()),
        "최소 어절 수": int(df["word_count"].min()),
        "최대 어절 수": int(df["word_count"].max()),
    }
    return summary
```

```python
# =========================================================
# 5. 전체 실행
# =========================================================

cleaned_dfs = {}
sampled_dfs = {}
before_after_stats = []

for code, (topic_name, df) in topic_map.items():
    print(f"\n===== {topic_name} ({code}) 처리 시작 =====")
    print(f"[원본] 문장 수: {len(df):,}")
    
    cleaned = clean_dataframe(df, text_col="text")
    cleaned_dfs[code] = cleaned
    
    print(f"[정제 후] 문장 수: {len(cleaned):,}")
    
    sampled = stratified_sample_by_length(
        cleaned,
        target_n=TARGET_N,
        text_col="text",
        random_state=RANDOM_STATE
    )
    sampled_dfs[code] = sampled
    
    print(f"[샘플링 후] 문장 수: {len(sampled):,}")
    
    before_after_stats.append({
        "주제코드": code,
        "주제명": topic_name,
        "원본 문장 수": len(df),
        "정제 후 문장 수": len(cleaned),
        "최종 샘플 문장 수": len(sampled),
        "정제 후 총 어절 수": int(cleaned["word_count"].sum()),
        "최종 샘플 총 어절 수": int(sampled["word_count"].sum()),
        "최종 샘플 평균 어절 수": round(sampled["word_count"].mean(), 2)
    })

stats_df = pd.DataFrame(before_after_stats)
print("\n===== 전체 요약 =====")
print(stats_df)
```
